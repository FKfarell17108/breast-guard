## Standarisasi Fitur Antar Dataset

```python
import pandas as pd
import numpy as np
```

```python
def standardize_dataset(df, source_name=""):
    rename_map = {
        "BMI": "imc", "bmi": "imc", "imc": "imc",
        "Weight": "weight", "weight": "weight",
        "Age": "age", "age": "age",
        "Menarche": "menarche", "menarche": "menarche",
        "Menopause": "menopause", "menopause": "menopause",
        "AgeFirstBirth": "agefirst", "agefirst": "agefirst",
        "Children": "children", "children": "children",
        "Breastfeeding": "breastfeeding", "Breast": "breastfeeding", "breastfeeding": "breastfeeding",
        "RelativesBC": "nrelbc", "nrelbc": "nrelbc",
        "Exercise": "exercise", "exercise": "exercise",
        "Alcohol": "alcohol", "alcohol": "alcohol",
        "Tobacco": "tobacco", "tobacco": "tobacco",
        "Allergies": "allergies", "allergies": "allergies",
        "Cancer": "cancer", "cancer": "cancer"
    }
    df = df.rename(columns=rename_map)

    binary_cols = ["menopause","breastfeeding","exercise","alcohol","tobacco","cancer"]
    for col in binary_cols:
        if col in df.columns:
            df[col] = df[col].replace({"Yes":1,"No":0,"Y":1,"N":0,"yes":1,"no":0})

    if "allergies" in df.columns:
        df["allergies"] = df["allergies"].apply(lambda x: 0 if (pd.isna(x) or str(x).lower()=="no") else 1)

    if "nrelbc" in df.columns:
        df["nrelbc"] = df["nrelbc"].apply(lambda x: 0 if (pd.isna(x) or str(x).lower()=="none") else 1)

    for col in ["age","menarche","agefirst","children"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    return df
```

```python
datasets = [df1, df2, df3, df4, df5, df6, df7]

standardized_datasets = []

for i, df in enumerate(datasets):
    df_standardized = standardize_dataset(df, source_name=f"df{i+1}")
    standardized_datasets.append(df_standardized)
```

```python
neg, pos = (y_train == 0).sum(), (y_train == 1).sum()
scale_pos_weight = neg / pos
print(f"scale_pos_weight = {scale_pos_weight:.2f}")
```

### Model

```python
# Model awal (baseline)
xgb_clf = xgb.XGBClassifier(
    n_estimators=200,
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    use_label_encoder=False,
    eval_metric="logloss"
)
```

### Train Model

```python
xgb_clf.fit(X_train, y_train)
```

```python
import numpy as np
import pandas as pd

def predict_risk(model, scaler, user_input):
    # Pastikan urutan sesuai dengan features
    features = [
        "age", "menarche", "menopause", "agefirst", "children",
        "breastfeeding", "nrelbc", "imc", "weight",
        "alcohol", "tobacco", "allergies"
    ]

    # Buat DataFrame dari input user
    df_input = pd.DataFrame([user_input], columns=features)

    # Scaling pakai scaler yang sudah fit di training
    df_input_scaled = scaler.transform(df_input)

    # Prediksi
    proba = model.predict_proba(df_input_scaled)[0][1]
    pred = model.predict(df_input_scaled)[0]

    return pred, proba
```

```python
# Input contoh
user_input = [
    45, 12, 1, 24, 2,   # age, menarche, menopause, agefirst, children
    1, 0, 28, 65,       # breastfeeding, nrelbc, imc, weight
    0, 0, 0          # alcohol, tobacco, allergies
]

pred, proba = predict_risk(xgb_clf, scaler, user_input)
print(f"Prediction: {pred}, Probability: {proba:.4f}")
```
